# Creating Dimensions

### Teams Dimensions

```python
def team_dim(data):
    # teams = sorted(pd.unique(data[['HomeTeam', 'AwayTeam']].values.ravel()))
    teams = sorted(data['Club'].unique())
    team_df = pd.DataFrame(teams, columns=['Name'])
    team_df['id'] = range(1, len(team_df) + 1)
    return team_df
```

```python
team_df = team_dim(df)
```

### Nationality Dimension

```python
def nationality_dim(data):
    country = sorted([x for x in data['Nationality'].unique() if not isinstance(x, float) or not math.isnan(x)])
    df = pd.DataFrame(country, columns=['Name'])
    df['id'] = range(1, len(df) + 1)
    return df
```

```python
nationality_dim(df)
```

### Position Dimension

```python
def pos_dim(data):
    pos = data['Position'].unique()
    df = pd.DataFrame(pos, columns=['Position'])
    df['id'] = range(1, len(df) + 1)
    return df
```

```python
pos_dim(df)
```

### Referee Dimension

```python
def ref_dim(data):
    refs = data['Referee'].unique()
    ref_df = pd.DataFrame(refs, columns=['Name'])
    ref_df['id'] = range(1, len(ref_df) + 1)
    return ref_df
```
